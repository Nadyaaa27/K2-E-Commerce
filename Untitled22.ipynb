```python
# import library
import pandas as pd
import numpy as np

# import data
star = pd.read_csv ('https://github.com/YBIFoundation/Dataset/raw/main/Stars.csv')

star.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Membuat objek LabelEncoder
label_encoder = LabelEncoder()

# Mengubah semua kolom kategorikal menjadi numerik
for column in star.columns:
    if star[column].dtype == 'object':
        star[column] = label_encoder.fit_transform(star[column])
```

```python
star.head()
```

```python
null_c = star.isnull().sum()
any_null = star.isnull().any(axis=0)
print('kolom  | baris kosong  | banyak baris kosong')
print('*'*50)
for column in star.columns:
  print(f"{column:<19}  | {any_null[column]:<19}  |   {null_c[column]:<19} ")
```

```python
X = star [['Temperature (K)', 'Luminosity (L/Lo)', 'Radius (R/Ro)',
       'Absolute magnitude (Mv)', 'Star type', 'Star category', 'Star color']]
y = star ['Spectral Class']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.22, random_state=123)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# standarisasi nilai-nilai dari dataset
scaler = StandardScaler()
scaler.fit(star)
X = scaler.transform(star)

# membuat objek SVC dan memanggil fungsi fit untuk melatih model
clf = SVC()
clf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score
import numpy as np


cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=42)
scores = cross_val_score(clf, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
print('Akurasi: %.3f' % np.mean(scores))
```
